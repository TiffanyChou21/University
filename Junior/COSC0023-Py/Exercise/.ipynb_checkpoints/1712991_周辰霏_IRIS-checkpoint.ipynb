```python
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
pd.set_option('display.max_rows',None)
iris1=pd.read_csv('iris.csv',names=['slength','swidth','plength','pwidth','label'])
iris=np.loadtxt('iris.csv',delimiter=",")
iris
```

```python
iris1.drop_duplicates()   #不是很理解每列排序去重，所以总体和对于每列的单独去重都做了    #本行全部去重
```

```python
iris1.sort_index(by='slength')    #排序去重第一列
iris1.drop_duplicates(['slength'])
```

```python
iris1.sort_index(by='swidth')
iris1.drop_duplicates(['swidth'])#排序去重第二列
```

```python
iris1.sort_index(by='plength')
iris1.drop_duplicates(['plength'])#排序去重第三列
```

```python
iris1.sort_index(by='pwidth')
iris1.drop_duplicates(['pwidth'])#排序去重第四列
```

```python
# 每列和、累积和、均值、标准差、方差、最小值、最大值
print(iris1.describe())
```

```python
print("每列和:\n"+str(iris1.apply(sum)))#每列和
```

```python
iris1.cumsum() #累计和
```

```python
print("均值:\n"+str(iris1.mean()))
```

```python
print("标准差:\n"+str(iris1.std()))
```

```python
print("方差:\n"+str(iris1.var()))
```

```python
print("最小值:\n"+str(iris1.min()))
```

```python
print("最大值:\n"+str(iris1.max()))
```
